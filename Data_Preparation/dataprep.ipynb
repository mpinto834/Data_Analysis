```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Sample Dataset Creation
def create_sample_dataset():
    data = {
        "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan", "Fiona", "George", "Hannah"],
        "Age": [25, 30, 22, 28, 35, 27, 32, 29],
        "Gender": ["Female", "Male", "Male", "Female", "Male", "Female", "Male", "Female"],
        "City": ["New York", "San Francisco", "Chicago", "Los Angeles", "Houston", "Miami", "Seattle", "Boston"]
    }
    df = pd.DataFrame(data)
    df.to_csv("sample_dataset.csv", index=False)

def load_dataset(file_path):
    df = pd.read_csv(file_path)
    return df

create_sample_dataset()
df = load_dataset("sample_dataset.csv")
```

```python
#Real Dataset
df = pd.read_csv("../dataset/train.csv")
```

```python
#Display the first few rows of the dataset
df.head()

# Get summary information
df.info()

# Descriptive statistics
print(df.describe())

# table of categorical variables
print(df['Name'].value_counts())

# Checking for missing values
print(df.isnull().sum())
```

```python
# Visualizing Age Distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=5, color='cornflowerblue', alpha=0.85, edgecolor='black')
plt.title('Age Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig('age_distribution.png')
plt.show()
```
