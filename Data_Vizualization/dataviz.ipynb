```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Load the dataset
df = pd.read_csv("../Data_Quality/cleaned_dataset.csv")
```

```python

# Set Seaborn style
sns.set(style="whitegrid")
sns.set_palette("Set2")
```

```python

# Bar Plot - Categorical Variable (Sex)
sns.countplot(data=df, x="Sex")
plt.title("Count by Gender")
plt.savefig("barplot.png")
plt.show()
```

```python

# Histogram - Numeric Variable (Age)
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Distribution of Age")
plt.savefig("AgeDistribution.png")
plt.show()
```

```python

# KDE Plot - Fare
sns.kdeplot(data=df, x="Fare", fill=True)
plt.title("Fare Distribution")
plt.savefig("Fare.png")
plt.show()
```

```python

# Box Plot - Fare by Pclass
sns.boxplot(data=df, x="Pclass", y="Fare")
plt.title("Fare by Passenger Class")
plt.savefig("Fare_by_Pclass.png")
plt.show()
```

```python

# Violin Plot - Age by Pclass
sns.violinplot(data=df, x="Pclass", y="Age")
plt.title("Age Distribution by Passenger Class")
plt.savefig("age_distribution_by_pclass.png")
plt.show()
```

```python

# Strip Plot - Age by Embarked
sns.stripplot(data=df, x="Embarked", y="Age", jitter=True)
plt.title("Age by Port of Embarkation")
plt.savefig("Age_by_PortofEmbarkation.png")
plt.show()
```

```python

# Scatter Plot - Age vs. Fare
sns.scatterplot(data=df, x="Age", y="Fare")
plt.title("Age vs Fare")
plt.savefig("Age_vs_Fare.png")
plt.show()
```

```python

# Scatter Plot with Hue - Age vs. Fare by Sex
sns.scatterplot(data=df, x="Age", y="Fare", hue="Sex")
plt.title("Age vs Fare by Gender")
plt.savefig("Age_vs_Fare_by_Gender.png")
plt.show()
```

```python

# Regression Plot - SibSp vs Fare
sns.regplot(data=df, x="SibSp", y="Fare", scatter_kws={'alpha':0.3})
plt.title("Number of Siblings/Spouses vs Fare with Regression Line")
plt.savefig('regplot.png')
plt.show()
```

```python

# FacetGrid: Histogram of Age by Sex
g = sns.FacetGrid(df, col="Sex")
g.map_dataframe(sns.histplot, x="Age", bins=15, kde=True)
g.fig.suptitle("Distribution of Age by Gender", y=1.03)
plt.savefig("Hist_age_sex.png")
plt.show()
```

```python

# FacetGrid: Scatter plot of Age vs Fare by Pclass
g = sns.FacetGrid(df, col="Pclass")
g.map_dataframe(sns.scatterplot, x="Age", y="Fare")
g.fig.suptitle("Age vs Fare by Passenger Class", y=1.03)
plt.savefig("Age_vs_Fare_by_Pclass.png")
plt.show()
```

```python

# Catplot: Boxplot of Fare by Sex across Embarked ports
sns.catplot(data=df, x="Sex", y="Fare", col="Embarked", kind="box")
plt.subplots_adjust(top=0.8)
plt.suptitle("Fare by Gender and Port of Embarkation")
plt.savefig("catplot.png")
plt.show()
```

```python

# Survival Analysis
# Bar Plot - Survival Rate
sns.barplot(data=df, x="Survived", y=df["Survived"].value_counts(normalize=True), palette="viridis")
plt.title("Overall Survival Rate")
plt.xlabel("Survived (0=No, 1=Yes)")
plt.ylabel("Percentage")
plt.show()
```

```python

# Bar Plot - Survival Rate by Sex
sns.barplot(x="Sex", y="Survived", data=df, errorbar=None, palette="coolwarm")
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.savefig("survival_rate_by_sex.png") 
plt.show()
```

```python

# Bar Plot - Survival Rate by Pclass
sns.barplot(x="Pclass", y="Survived", data=df, errorbar=None, palette="viridis")
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.savefig("survival_rate_by_pclass.png") 
plt.show()
```
